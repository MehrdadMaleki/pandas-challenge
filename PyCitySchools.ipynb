```python
import pandas as pd

school_data_to_load ='../Resources/schools_complete.csv'
student_data_to_load = '../Resources/students_complete.csv'

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"]) 
school_data_complete.head()
```

```python
# District Summary

totalnumberofschools=school_data_complete["school_name"].nunique()
Calculatethetotalnumberofstudents=school_data_complete["student_name"].count()
Calculatethetotalbudget=school_data_complete.groupby(["school_name"])["budget"].first().sum()
AverageMathScore=school_data_complete["math_score"].mean()
AverageReadingScore=school_data_complete["reading_score"].mean()
greater70=school_data_complete.loc[school_data_complete["math_score"]>=70,: ]
greater70s=len(greater70)
PassingMath=(greater70s/Calculatethetotalnumberofstudents)*100
greater70r=school_data_complete.loc[school_data_complete["reading_score"]>=70 ,:]
greater70rs=len(greater70r)
PassingReading=(greater70rs/Calculatethetotalnumberofstudents)*100
overall=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]
Overall1=len(overall)
OverallPassing=(Overall1/Calculatethetotalnumberofstudents)*100

DistrictSummary= pd.DataFrame({"Total Schools" : [totalnumberofschools],"Total Students":[Calculatethetotalnumberofstudents],
                             "Total Budget":[Calculatethetotalbudget],"Average Math Score": [AverageMathScore],
                            "Average Reading Score":[AverageReadingScore],"% Passing Math":[PassingMath],
                             "% Passing Reading":[PassingReading],"% Overall Passing":[OverallPassing] })

DistrictSummary["Total Budget"]=DistrictSummary["Total Budget"].astype(float).map("${:,}".format)
DistrictSummary

```

```python
#School Summary

ss1=school_data_complete.groupby(["school_name"]).agg({
         'type': 'first',
         'student_name': 'count',   
         'budget': 'first',
        'math_score': 'mean',
         'reading_score':'mean',
        
    })


ss2 = greater70.groupby(["school_name"])["math_score"].count().reset_index()

ss3 = greater70r.groupby(["school_name"])["reading_score"].count().reset_index()

ss4 = overall.groupby(["school_name"])["student_name"].count().reset_index()

ss1['Per Student Budget'] = round( ss1['budget']  / ss1['student_name'],2)


ss5 = pd.merge(ss1,ss2, on="school_name" )
ss6 = pd.merge(ss3,ss4, on="school_name" )
ss7 = pd.merge(ss5,ss6, on="school_name" )

ss7['math_score_y'] = round( (ss7['math_score_y']  / ss7['student_name_x'])*100 ,2)
ss7['reading_score_y'] = round( (ss7['reading_score_y']  / ss7['student_name_x'])*100 ,2)
ss7['student_name_y'] = round( (ss7['student_name_y']  / ss7['student_name_x'])*100 ,2)


SchoolSummary=ss7.rename(columns={"type":"School Type", "student_name_x":"Total Students",
                                  "budget":"Total School Budget","math_score_x":"Average Math Score",
                                  "reading_score_x":"Average Reading Score","math_score_y":"% Passing Math",
                                  "reading_score_y":"% Passing Reading","student_name_y":"% Overall Passing"})

SchoolSummary["Total School Budget"]=SchoolSummary["Total School Budget"].astype(float).map("${:,}".format)
SchoolSummary["Per Student Budget"]=SchoolSummary["Per Student Budget"].astype(float).map("${:,}".format)


SchoolSummary= SchoolSummary.set_index('school_name')

SchoolSummary= SchoolSummary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                             "Average Math Score","Average Reading Score","% Passing Math", "% Passing Reading",
                            "% Overall Passing" ]]

SchoolSummary
```

```python
# Top Performing Schools (By % Overall Passing)

TopPerformingSchools=SchoolSummary.sort_values(by='% Overall Passing', ascending=False).head()
TopPerformingSchools
```

```python
# Bottom Performing Schools (By % Overall Passing)

BottomPerformingSchools=SchoolSummary.sort_values(by='% Overall Passing', ascending=True).head()
BottomPerformingSchools
```

```python
# Scores by School Spending

school_data_complete["count Student"] = school_data_complete.groupby(["school_name"])["Student ID"].transform('count')

school_data_complete["Spending Per"] = school_data_complete["budget"]/school_data_complete["count Student"]


bins = [0, 584, 630, 645, 676]

label_Spending =["<584", "585-629 ","630-644 ", "645-675"]
school_data_complete["Spending Ranges (Per Student)"]=pd.cut(school_data_complete["Spending Per"],
                                                             bins,labels=label_Spending, right=False)



sr1=school_data_complete.groupby(["Spending Ranges (Per Student)"]).agg({'math_score': 'mean','reading_score':'mean'})


greater70=school_data_complete.loc[school_data_complete["math_score"]>=70,: ]
greater70r=school_data_complete.loc[school_data_complete["reading_score"]>=70 ,:]
overall=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]


sr2 = greater70.groupby(["Spending Ranges (Per Student)"])["math_score"].count().reset_index()

sr3 = greater70r.groupby(["Spending Ranges (Per Student)"])["reading_score"].count().reset_index()

sr4 = overall.groupby(["Spending Ranges (Per Student)"])["Student ID"].count().reset_index()

sr8 = school_data_complete.groupby(["Spending Ranges (Per Student)"])["Student ID"].count().reset_index()

sr5 = pd.merge(sr1,sr2, on="Spending Ranges (Per Student)" )
sr6 = pd.merge(sr3,sr4, on="Spending Ranges (Per Student)" )
sr7 = pd.merge(sr5,sr6, on="Spending Ranges (Per Student)" )
sr9 = pd.merge(sr7,sr8, on="Spending Ranges (Per Student)" )


sr9['% Passing Math'] = round( (sr9['math_score_y']  / sr9['Student ID_y'])*100 ,2)
sr9['% Passing Reading'] = round( (sr9['reading_score_y']  / sr9['Student ID_y'])*100 ,2)
sr9['% Overall Passing'] = round( (sr9['Student ID_x']  / sr9['Student ID_y'])*100 ,2)

sr9=sr9.rename(columns={"math_score_x":"Average Math Score",
                        "reading_score_x":"Average Reading Score"})

ScoresbySchoolSpending= sr9[["Spending Ranges (Per Student)","Average Math Score","Average Reading Score",
                            "% Passing Math", "% Passing Reading","% Overall Passing" ]]

ScoresbySchoolSpending
```

```python
# Scores by School Size

bins = [0, 999, 1999, 5000]

label_size =["small(<1000)", "Medium(1000-2000)","Large(2000-5000)"]
school_data_complete["school size"]=pd.cut(school_data_complete["size"],
                                                             bins,labels=label_size, right=True)



ssr1=school_data_complete.groupby(["school size"]).agg({'math_score': 'mean','reading_score':'mean'})


ssrgreater70=school_data_complete.loc[school_data_complete["math_score"]>=70,: ]
ssrgreater70r=school_data_complete.loc[school_data_complete["reading_score"]>=70 ,:]
ssroverall=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]


ssr2 = ssrgreater70.groupby(["school size"])["math_score"].count().reset_index()

ssr3 = ssrgreater70r.groupby(["school size"])["reading_score"].count().reset_index()

ssr4 = ssroverall.groupby(["school size"])["student_name"].count().reset_index()

ssr8 = school_data_complete.groupby(["school size"])["student_name"].count().reset_index()

ssr5 = pd.merge(ssr1,ssr2, on="school size" )
ssr6 = pd.merge(ssr3,ssr4, on="school size" )
ssr7 = pd.merge(ssr5,ssr6, on="school size")
ssr9 = pd.merge(ssr7,ssr8, on="school size" )


ssr9['% Passing Math'] = (ssr9['math_score_y']  / ssr9['student_name_y'])*100 
ssr9['% Passing Reading'] = (ssr9['reading_score_y']  / ssr9['student_name_y'])*100 
ssr9['% Overall Passing'] =  (ssr9['student_name_x']  / ssr9['student_name_y'])*100 

ssr9=ssr9.rename(columns={"math_score_x":"Average Math Score",
                        "reading_score_x":"Average Reading Score"})

ScoresbySchoolSize= ssr9[["school size","Average Math Score","Average Reading Score",
                            "% Passing Math", "% Passing Reading","% Overall Passing" ]]

ScoresbySchoolSize
```

```python
# Scores by School Type

tssr1=school_data_complete.groupby(["type"]).agg({'math_score': 'mean','reading_score':'mean'})


tssrgreater70=school_data_complete.loc[school_data_complete["math_score"]>=70,: ]
tssrgreater70r=school_data_complete.loc[school_data_complete["reading_score"]>=70 ,:]
tssroverall=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]


tssr2 = tssrgreater70.groupby(["type"])["math_score"].count().reset_index()

tssr3 = tssrgreater70r.groupby(["type"])["reading_score"].count().reset_index()

tssr4 = tssroverall.groupby(["type"])["student_name"].count().reset_index()

tssr8 = school_data_complete.groupby(["type"])["student_name"].count().reset_index()

tssr5 = pd.merge(tssr1,tssr2, on="type" )
tssr6 = pd.merge(tssr3,tssr4, on="type" )
tssr7 = pd.merge(tssr5,tssr6, on="type")
tssr9 = pd.merge(tssr7,tssr8, on="type" )


tssr9['% Passing Math'] = (tssr9['math_score_y']  / tssr9['student_name_y'])*100
tssr9['% Passing Reading'] = (tssr9['reading_score_y']  / tssr9['student_name_y'])*100 
tssr9['% Overall Passing'] = (tssr9['student_name_x']  / tssr9['student_name_y'])*100

tssr9=tssr9.rename(columns={"math_score_x":"Average Math Score",
                        "reading_score_x":"Average Reading Score"})

ScoresbySchooltype= tssr9[["type","Average Math Score","Average Reading Score",
                            "% Passing Math", "% Passing Reading","% Overall Passing" ]]

ScoresbySchooltype
```
